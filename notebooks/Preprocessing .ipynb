## Preprocessing on Parts column <a id='preprocessingOnPartCol'></a>

```python
# Take A Copy form origin dataframe to preforme any prerpocessing on Part Columns
parts_df = ar_df.copy()
```

```python
# Repalce the letters to the one unique char
def ReplaceChars(x):
    x = str(x)
    x = x.replace('ة','ه')
    x = x.replace('إ','ا')
    x = x.replace('أ','ا')
    x = x.replace('عضمه','عظمه')
    x = x.replace('ظفيره','ضفيره') 
    x = re.sub('\s+', ' ', x) # If there more than 1 space replace it with one space
    return x 
```

```python
# Split the data
def SplitText(x):
    x = str(x)
    x = re.split('\s',x)
    return x
```

```python
# Apply replcae chars function to reuduce the unique values as much possiable 
parts_df['Parts']=parts_df['Parts'].apply(ReplaceChars)
```

```python
# Apply Split text function to access to the part easily later
parts_df['ListParts']=parts_df['Parts'].apply(SplitText)
```

```python
# Save all unique parts which is more than 60k values
allParts = parts_df['Parts'].unique()
```

### Get the part names
So, we decide here to get the first element in list part array to make it as the main part name an if the element has less than two char we will go to the second element 

```python
# Get first item in parts list
def PartTypeName_ByFirstIndex(x):
    if len(x)>1:
        for i,s in enumerate(x):
            if len(s)<2:
                continue
            else:
                return x[i]
    else:
        return x[0]
```

Now we add new column for the part name and apply the previous function to take the name of first part in the list

```python
parts_df['PartName'] = parts_df['ListParts'].apply(PartTypeName_ByFirstIndex)
```

### Get the part positon names
Most of car part has position so we extract it from the list base on the funciton in bottom 

```python
def PartPosition(x):
    
    # Declear the main part position in list
    positionList=['امامي','خلفي','يمين','يسار']

    # Declear boolean variable to check there is previous word in the final string or not
    flag = False

    # Declear count to check number of part position 
    count= 0

    # Declear final part poistion string
    position = ''

    # Check if the curernt list greater than 1 elemnt than maybe have a postion else there no possiable to find part position
    if len(x)>1:
        for s in x:
            # Compare current element with postion list
            for pos in positionList:
                # Check if the current string has similarity with more than 90 % then add it as new position 
                if jf.jaro_similarity(s,pos)>0.9:
                    # This word in the next condtion have 90% similarty with 'خلفي' but actual is not the positoin we want
                    if s == 'خلف':
                        continue
                    # This check if postion have 2 acutal postion so we will stop. Beacuse we have just 4 dicrction and 4 sub direction
                    if count == 2:
                        break
                    # If there previous string that add space to put new word 
                    if flag :
                        position += ' '
                    
                    # Now we add the actual positon 
                    position += pos
                    flag = True
                    count+=1

                # Another Check if the current string has similarity with more than 80 % then add it as new position
                elif jf.jaro_similarity(s,pos)>0.8:
                    if count == 2:
                        break
                    if flag :
                        position += ' '
                    position += pos
                    flag = True
                    count+=1

    # If the position is empty then there is no positon so will fill it with undefiend string
    if position == '':
        position = 'غير محدد'

    '''
    Now we have more than 4 main postion and 4 sub position (خلفي, امامي,امامي يمين,امامي يسار,خلفي يسار ,خلفي يمين, يمين, يسار)  
    But some times the result will be repeated for example (أمامي أمامي), or there problem with order (يمين أمامي) the main position is (أمامي) and not (يمين)
    So will be change the orders and remove the repeated postion or if there two main positon in same string (أمامي خلفي) so we will take the first in order    
    '''
    # If the position is defined enter this conditon
    if position != 'غير محدد':
        mainPosition=['خلفي','امامي']

        # If the postion string with more than 1 word then we will check
        check=position.split(' ')
        if len(check)>1:
            # Declear new positon string 
            newposition=''
            
            # If the positon is repeated we will take the first one
            if check[0]==check[1]:
                newposition += check[0]
                return newposition

            # If there two main position we will take the first one 
            if check[0]=='امامي' and check[1]=='خلفي':
                return check[0]
            elif check[1]=='امامي' and check[0]=='خلفي':
                return check[0]

            # If the main postion is in second order make it first one 
            if check[1] in mainPosition:
                newposition += (check[1]+' '+check[0])
                return newposition

            # If there two submain positions we will take the first one 
            if check[0]=='يمين' and check[1]=='يسار':
                return check[0]
            elif check[1]=='يمين' and check[0]=='يسار':
                return check[0]


    return position
```

```python
parts_df['PartPosition'] = parts_df['ListParts'].apply(PartPosition)
```

## Preprocessing on PartName (clustring)
So, after we take the first index part that leave us with 2000 unique parts but also there repeteing and similarity between them, so we find way to reduce the unique values by use jaro similarty to cluster the parts name. 

### Apply it first on dictoinary 

```python
# Get all unique part name in list
uniqueParts = parts_df['PartName'].value_counts().index.tolist()

# Get all smiliraty on dictonary
sames={}

# To prevent to check visited values
visitedi=[]
visitedj=[]

# Check each unique value with whole unique parts if find the smiliarty add to sames dictonary
for (i) in uniqueParts:
    # If the part were found it's similarity then will skip it   
    if i in visitedj:
        continue
    
    # Add the current iterate in i list to doesn't visited again 
    visitedi.append(i)

    for j in uniqueParts:
        # If current value in i list equal the current value in j list thats mean it's same value so we will skip it
        if i == j:
            continue

        # Also if the part were found it's similarity then will skip it 
        if j in visitedi:
            continue

        # Convert currents values in two lists to string
        i = str(i)
        j = str(j)

        # If current value in i list has similarity more than 74% than add it to sames dictonary and add the 
        if jf.jaro_similarity(i,j)>0.74:
            try:
                if sames[i]:
                    sames[i].append(j)
            except:
                sames[i]=[j]

            # Add the current iterate in j list to doesn't visited again 
            visitedj.append(j)
sames
```

### Apply it on actual dataframe

```python

count = 0 
visitedi=[]
visitedj=[]
# Iterate on each Part Name 
for (i) in uniqueParts:
    # If the part were found it's similarity then will skip it   
    if i in visitedj:
        continue
    
    # Add the current iterate in list 
    visitedi.append(i)
    for j in uniqueParts:
        # If the part is itself the we will skip it 
        if i == j:
            continue
        # Also if the part were found it's similarity then will skip it 
        if j in visitedi:
            continue

        i = str(i)
        j = str(j)
        # If the similarity is more than 74% than we replace the main value on dataframe 
        if jf.jaro_similarity(i,j)>0.74:
            parts_df['PartName'].replace({j:i}, inplace=True)
            visitedj.append(j)
```

```python
# Get all columns name into list and add our new column key which is c_id
df_keys=test_df.keys().to_list()
df_keys.append('c_id')

# Initalize the count to use it as new id 
count = 0
# Initlize previous_row variable and will contains the previous row inforamiton on the dataframe
previous_row = None

# This is boolean var to know if we find the previous row it's same with current row 
flag = False

# Initlaize Dictonary to save all information on new dataframe
new_df = {}
for x in df_keys:
    new_df[x]={}

# Start to loop on the whole dataframe    
for row in test_df.itertuples():
    
    # current id will equal the count
    id = count

    # Check if the current row equal the previous row expect the partsname than make the current row id equal the previous id (count-1) then make the flag true
    try:
        if (previous_row.TotalCost == row.TotalCost) and (previous_row.CloseTime ==row.CloseTime) and (previous_row.CarModel == row.CarModel) and (previous_row.ManufactureYear == row.ManufactureYear) and (previous_row.CarColor == row.CarColor) and (previous_row.Area == row.Area):
            id = (count-1) 
            flag = True
    except AttributeError:
        pass
        
    # Append the current row data to new df dictonary    
    for key in df_keys:
        if key == 'c_id':
            new_df['c_id'][row.Index]= id
            continue
        # This line to make the x variable equal the row key 
        exec(f"x = row.{key}")
        new_df[key][row.Index]= x
        
    previous_row = row
    if flag:
        flag=False
        continue
    else:
        count+=1
    
```
